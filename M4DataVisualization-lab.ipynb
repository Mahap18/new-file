# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Open a connection to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve ConvertedComp data from the master table
QUERY = """
SELECT ConvertedComp
FROM master
WHERE ConvertedComp IS NOT NULL
"""
# Execute the query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the connection to the database
conn.close()

# Plot the histogram for the 'ConvertedComp' column
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size
plt.hist(df['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')  # dropna() not needed as we filtered out NaN values in the query
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp (Annual Salary in USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


```

### Box Plots


Plot a box plot of `Age.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Open a connection to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve Age data from the master table
QUERY = """
SELECT Age
FROM master
WHERE Age IS NOT NULL
"""
# Execute the query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the connection to the database
conn.close()

# Plot the box plot for the 'Age' column
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size
sns.boxplot(x=df['Age'], color='skyblue')
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()


```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Open a connection to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve Age and WorkWeekHrs data from the master table
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL
"""
# Execute the query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the connection to the database
conn.close()

# Plot the scatter plot for 'Age' vs. 'WorkWeekHrs'
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df, color='skyblue')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()


```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Open a connection to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve WorkWeekHrs, CodeRevHrs, and Age data from the master table
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL AND Age IS NOT NULL
"""
# Execute the query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the connection to the database
conn.close()

# Create a bubble plot where Age determines the bubble size
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size
plt.scatter(
    x=df['WorkWeekHrs'],       # x-axis: WorkWeekHrs
    y=df['CodeRevHrs'],        # y-axis: CodeRevHrs
    s=df['Age']*10,            # Bubble size is proportional to Age (scaled by 10 for better visibility)
    alpha=0.5,                 # Transparency of the bubbles
    color='skyblue',           # Color of the bubbles
    edgecolors='black'         # Add border color to the bubbles
)

# Add titles and labels
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs with Age as Bubble Size')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')

# Display the plot
plt.grid(True)
plt.show()


```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Open a connection to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve relevant columns for the age group 30-35
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
# Execute the query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the connection to the database
conn.close()

# Calculate the median values of WorkWeekHrs and CodeRevHrs
median_workweekhrs = df['WorkWeekHrs'].median()
median_coderevhrs = df['CodeRevHrs'].median()

# Prepare data for plotting (stacked bar chart)
median_values = [median_workweekhrs, median_coderevhrs]

# Labels for the bars
labels = ['WorkWeekHrs', 'CodeRevHrs']

# Plotting the stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(labels, median_values, color=['skyblue', 'lightcoral'])

# Adding title and labels
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.ylabel('Median Hours')
plt.show()


```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Open a connection to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve relevant columns for the age group 45-60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60 AND ConvertedComp IS NOT NULL
"""
# Execute the query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the connection to the database
conn.close()

# Group by Age and calculate the median ConvertedComp for each age
median_salary_by_age = df.groupby('Age')['ConvertedComp'].median()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_salary_by_age.index, median_salary_by_age.values, marker='o', color='b', linestyle='-', linewidth=2, markersize=6)

# Adding title and labels
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp (USD)')
plt.grid(True)
plt.show()


```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Open a connection to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the MainBranch column
QUERY = """
SELECT MainBranch
FROM master
WHERE MainBranch IS NOT NULL
"""
# Execute the query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the connection to the database
conn.close()

# Count the occurrences of each unique value in the MainBranch column
mainbranch_counts = df['MainBranch'].value_counts()

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue', edgecolor='black')

# Adding title and labels
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(True, axis='x')

# Display the plot
plt.show()


```
